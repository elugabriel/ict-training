##  1. What is Data Science?

Data Science is an interdisciplinary field that extracts meaningful insights from data using **mathematics**, **statistics**, **programming**, and **domain knowledge**.

üîç It involves:
- Data collection
- Data cleaning & preprocessing
- Exploratory data analysis (EDA)
- Building and evaluating models
- Communicating results


##  2. The CRISP-DM Framework

CRISP-DM stands for **Cross Industry Standard Process for Data Mining**. It is one of the most widely used data science workflows.


### CRISP-DM Stages:
1. **Business Understanding** ‚Äì Define the problem.
2. **Data Understanding** ‚Äì Gather and explore data.
3. **Data Preparation** ‚Äì Clean and format data.
4. **Modeling** ‚Äì Apply algorithms.
5. **Evaluation** ‚Äì Measure and refine model performance.
6. **Deployment** ‚Äì Deliver solution and monitor it.
